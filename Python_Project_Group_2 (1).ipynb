**#1) Create a simple Python program application that does the following:**

A) Prints "Hello User!"

 B) Then asks "What is your name?"

C) Then responds "Hello <user's name>"

D) Then asks: "What is your year of birth?"

E) Then ask: "Do you know how to drive?"

F) Then ask: "How old were you when you got your driving license" if the user replied "Yes" to the above question

G) Then outputs: "You got your driving license in <year> and you have been driving for <x> years" 

H) Then outputs: "You are <age> years old now but you still can't drive. You were supposed to get your license in <year>,<x> years ago at the age of 30 at the latest" If the user replied "No" and is over the age of 30

I) Then outputs: "You are <age> years old now and you should work on getting your driving license before you turn 30. You have <x> years to go until the year <year>" If the user replied "No" and is under the age of 30

```python
import datetime 

def get_valid_year(prompt):
    while True:
        try:
            year = int(input(prompt))
            if 1900 <= year <= datetime.datetime.now().year:
                return year
            else:
                print("Please enter a valid year between 1900 and the current year.")
        except ValueError:
            print("Please enter a valid number for the year.")

def get_valid_age(prompt, max_age):
    while True:
        try:
            age = int(input(prompt))
            if 0 <= age <= max_age:
                return age
            else:
                print(f"Please enter a valid age between 0 and {max_age}.")
        except ValueError:
            print("Please enter a valid number for age.")

# Print greeting
print("Hello User!")

# Ask for name
name = input("What is your name? ").strip()
print(f"Hello {name}")

# Ask for year of birth
birth_year = get_valid_year("What is your year of birth? ")

# Calculate current year and age
current_year = datetime.datetime.now().year
age = current_year - birth_year

# Ask about driving
while True:
    can_drive = input("Do you know how to drive? (Yes/No) ").lower().strip()
    if can_drive in ['yes', 'no']:
        break
    print("Please answer 'Yes' or 'No'.")

if can_drive == "yes":
    license_age = get_valid_age(f"How old were you when you got your driving license? (0-{age}) ", age)
    license_year = birth_year + license_age
    driving_years = current_year - license_year
    print(f"You got your driving license in {license_year} and you have been driving for {driving_years} years")
else:  # can_drive == "no"
    if age > 30:
        supposed_license_year = birth_year + 30
        years_ago = current_year - supposed_license_year
        print(f"You are {age} years old now but you still can't drive. You were supposed to get your license in {supposed_license_year}, {years_ago} years ago at the age of 30 at the latest")
    else:
        years_to_30 = 30 - age
        license_year = current_year + years_to_30
        print(f"You are {age} years old now and you should work on getting your driving license before you turn 30. You have {years_to_30} years to go until the year {license_year}")



```

#2) **Number Chain**
**In this activity, you will take user input and print out a series of numbers.**

**Instructions**

A) Using a while loop, ask the user "How many numbers?", and then print out a chain of numbers in increasing order, from 0 to the user-input number.

B) After the results have been printed, ask the user if they would like to continue .

C) If "y" is entered, keep the chain running by inputting a new number and starting a new count from 0 to the new user-input number.

D) If "n" is entered, exit the application.

**Bonus**

E) Rather than just displaying numbers starting from 0, have the numbers begin at the end of the previous chain.

```python
def number_chain():
    print("Welcome to the Number Chain Program!")
    continue_chain = "y"  # Initialize the loop control variable
    
    while continue_chain.lower() == "y":
        # Ask the user how many numbers they want
        try:
            count = int(input("How many numbers? "))
            if count < 0:
                print("Please enter a positive number.")
                continue
            
            # Print the chain of numbers
            print("Here is your number chain:")
            for num in range(count + 1):
                print(num, end=" ")
            print("\n")  # Add a new line after printing the chain
            
            # Ask if the user wants to continue
            continue_chain = input("Would you like to continue? (y/n): ").strip().lower()
        except ValueError:
            print("Invalid input. Please enter a valid number.")
    
    print("Thank you for using the Number Chain Program. Goodbye!")

# Run the program
number_chain()
```

**#5) Write a python program for a simple bank that accepts deposits, processes withdrawals and gives simple interest on deposits**

A) Your program should ask if the user wants to deposit or withdraw money

B) The program should perform the operation based on the user's response

C) Then, the user should be notified of the activity and the final balance once the activity has taken place. 

D) The program should also calculate the simple interest and the original principal from the current balance, assuming that the user's final balance has been  in the user's account for 5 years now and a 4% rate of interest.

E) Finally, the program should let the user know how much balance he/she currently has, how long it's been saved, and how much of the balance is accrued interest.

```python

# Simple Bank Program
#  Beginning of the account balance
balance = 0.0
# Interest parameters
rate_of_interest = 4.0  # Annual interest rate in percent
time = 5
# Function to calculate simple interest
def calculate_simple_interest(principal, rate, time):
    return (principal * rate * time) / 100
# Welcome New Customer
print("Welcome to the Simple Bank!")
while True:
    # Ask the user for the operation
    action = input("Do you want to Deposit(d), Withdraw(w), or Exit(e)? ").strip().lower()
    if action == "d":
        # if customer wants to deposit
        deposit_amount = float(input("Enter the amount you want to deposit: "))
        balance += deposit_amount
        print(f"Deposit successful! Your new balance is: ${balance:.2f}")
    elif action == "w":
        # if the customer wants to withdrawal
        withdrawal_amount = float(input("Enter the amount you want to withdraw: "))
        if withdrawal_amount <= balance:
            balance -= withdrawal_amount
            print(f"Withdrawal successful! Your new balance is: ${balance:.2f}")
        else:
            print("Insufficient balance. card decline.")
    elif action == "e":
        # Exit the program
        print("Thank you for choosing Simple Bank! Take your card")
        break
    else:
        print("Invalid option.Insert again.")
        continue
# Calculate simple interest
simple_interest = calculate_simple_interest(balance, rate_of_interest, time)
principal = balance - simple_interest
# Final balance and interest details
print("Account Summary:")
print(f"Final Balance: ${balance:.2f}")
print(f"Time Saved: {time} years")
print(f"Accrued Interest: ${simple_interest:.2f}")
print(f"Original Principal: ${principal:.2f}")

```

```python

candies = ["Snickers", "KitKat", "Twix", "M&Ms", "Skittles", "Milky Way", "Reese's"]
# Empty list to store the user's candy choices
candy_cart = []
# Print all candies with their index

print("Available Candies:")
for index, candy in enumerate(candies):
    print(f"[{index}] {candy}")
    
# Set the allowance (number of candies the user can select)

allowance = 5
print("You can select up to 5 candies!")

# Allow the user to select candies based on their allowance
for _ in range(allowance):
    choice = int(input("Enter the index of the candy you want: "))
    if 0 <= choice < len(candies):
        candy_cart.append(candies[choice])
        print(f"Added {candies[choice]} to your cart.")
    else:
        print("Invalid choice. Please select a valid candy index.")
# Print all the candies selected
print("Candies you selected:")
for candy in candy_cart:
    print(candy)
# BONUS: Allow unlimited candy selection until the user opts out
print("BONUS: Select as many candies as you want!")
candy_cart_bonus = []
while True:
    choice = int(input("Enter the index of the candy you want (or -1 to stop): "))
    if choice == -1:
        break
    elif 0 <= choice < len(candies):
        candy_cart_bonus.append(candies[choice])
        print(f"Added {candies[choice]} to your cart.")
    else:
        print("Invalid choice. Please select a valid candy index.")
# Print all the candies selected in the bonus round
print("Candies you selected in the bonus round:")
for candy in candy_cart_bonus:
    print(candy) 
```
